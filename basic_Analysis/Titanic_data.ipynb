## Basic Data Analysis using Pandas and NumPy:

Pandas and NumPy are popular Python libraries for data manipulation, analysis, and cleaning. Here's a brief overview of how they are used for basic data analysis

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data= pd.read_csv('./titanic.csv')
data.head()
```

### dropping the columns

```python
data.drop(columns=["PassengerId" , 'Cabin', 'Ticket'] , inplace = True)
data
```

## Group by

### genderwise survival

```python
sex_survived_sum =  data.groupby(['Sex'])['Survived'].sum()
sex_survived_sum
```

```python
pclass_survived_average = data.groupby(['Pclass' ,'Age'])['Survived'].mean()*100
pclass_survived_average
```

### quering data

```python
pclass_age_20 = data.query("Pclass == 3 & Age ==20")
pclass_age_20
```

```python
a= data.query("Age >= 60 & Sex == 'female' & Pclass == 3")
a
```

```python
b= data.query("Age <= 20 & Sex == 'female' & Pclass == 1")
b
```

## Pivot

```python
pivot1 = data.pivot_table(values= 'Fare', index='Sex' , columns= 'Pclass' , aggfunc='sum')
pivot1
```

```python
pivot2 = data.pivot_table(values='Survived' , index='Pclass' , columns='Sex', aggfunc='sum')
pivot2
#values must be categorical values, and index will be the row , columns will be columns
#the total of values is added.
```

# Replace

```python
data.Sex.replace({'male':0, 'female':1}, inplace=True)
data
```

```python
data.Survived.replace({0:'dead', 1:'alive'}, inplace= True)
data
```

```python
data.rename(columns={'Sex':'Gender'} , inplace=True)
data
```

# fillna

## pd.cut

```python
cutoff = [0,15,50,80]
labels = ['Child' , 'Adult','Old']
data['age_type'] = pd.cut(data.Age, bins=cutoff,labels=labels)
data.head(20)
```
