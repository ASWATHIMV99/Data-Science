# <center>Cuisine Dataset Analysis

## Data importing

```python
"""importing the necessary python libraries"""
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
"""importing the dataset"""

data= pd.read_csv('./Cuisine_rating.csv')
data.head()
```

## Data Preprocessing

```python
"""finding the data information such as column name null counts and data types"""
data.info()
```

```python
"""it shows column name of the data"""
data.columns
```

```python
"""Getng information of numeric columns about count , mean , median , mode , standard deviation , min , max , 25% , 50% , 75%  in a dataset"""
data.describe()
```

```python
"""taking ten sample values"""
data.sample(10)
```

```python
"""renaming the columns to avoid the spaces between them"""
data.rename(columns={'User ID':'User_ID','Marital Status':'Marital_Status','Alcohol ':'Alcohol','Food Rating':'Food_Rating','Service Rating':'Service_Rating','Overall Rating':'Overall_Rating','Often A S':'Often_A_S'}, inplace=True)
```

```python
"""checking if there any null values, finding sum of all null values"""
data.isnull().sum()
```

```python
"""dropping the duplicate values"""

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
"""calculating the age from given year of birth and adding a seperate column named Age"""

data['Age']=[(2022 - YOB) for YOB in data['YOB']]
data.head()
```

```python
"""dropping the unwanted columns"""

data.drop(columns=['Area code','YOB'] ,inplace= True)
data.head()
```

## Data Manipulation

### custom function to create location using dataframe

```python
"""creating a function to sort location vise, using def function"""


def sort_location(Location):
    data_location = data[data.Location== Location]
    data_location.drop(columns= 'Location',inplace=True)
    return data_location
```

```python
"""sorting the data by China_Town_NY location"""

China_Town_NY = sort_location('China Town, NY')
China_Town_NY
```
